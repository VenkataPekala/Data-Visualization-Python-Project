```python
import pandas as pd
```

```python
coffee=pd.read_csv("df_coffee_qlty.csv")
```

```python
coffee.head()
```

```python
coffee.drop(columns=["ID","Unnamed: 0"],axis=1,inplace=True)
```

```python
coffee.head()
```

```python
coffee.dtypes
```

```python
coffee.isnull().sum()
```

```python
coffee.dropna(inplace=True)
```

```python
correlation=coffee.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(15,6))
sns.heatmap(correlation,cmap="Blues",annot=True,)
plt.title("Heatmap Coffee Data")
plt.show()
```

```python
import plotly.express as px

fig = px.imshow(coffee.corr(), text_auto = True,color_continuous_scale = 'RdYlBu')

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(2)

# Data
correlation=coffee.corr()

#labels
xlab=list(coffee)

# Heat map
fig, ax = plt.subplots()
ax.imshow(coffee.corr())

# Add the labels
ax.set_xticks(np.arange(len(xlab)), labels = xlab)
ax.set_yticks(np.arange(len(xlab)), labels = xlab)

# Rotate the labels of the X-axis
plt.setp(ax.get_xticklabels(), rotation = 90,
         ha = "right", rotation_mode = "anchor")

# plt.show()
plt.show()
```
